# Sun Position Tracker

This is a Python program that tracks the position of the sun at regular intervals and calculates its azimuth and altitude for a given location.

## Features

- Calculates the sun's position (azimuth and altitude) based on date and location.
- Provides continuous tracking of the sun's position at a specified interval.
- Example usage provided for Victoria Square in Belfast, UK.

## Prerequisites

- Python 3.x
- `ephem` library (install using `pip install ephem`)

## Usage

1. Clone the repository:

   ```shell
   git clone https://github.com/your-username/sun-position-tracker.git
   ```

2. Navigate to the project directory:

   ```shell
   cd sun-position-tracker
   ```

3. Run the program:

   ```shell
   python sun_position_tracker.py
   ```

4. Adjust the latitude, longitude, start date, and duration according to your requirements.

## Configuration

- Latitude: The latitude coordinate of your location.
- Longitude: The longitude coordinate of your location.
- Start Date: The date and time from which you want to start tracking the sun's position.
- Duration: The duration (in seconds) for which you want to track the sun's position.
